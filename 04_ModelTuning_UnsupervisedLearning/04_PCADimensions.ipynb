```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X, y = make_blobs(n_samples=500, n_features=5, centers=3, cluster_std=1.5, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2','PC3'])
df_pca['label'] = y
```

```python
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='label', palette='Set2')
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], c=df_pca['label'], cmap='Set2', s=40)
legend1 = ax.legend(*scatter.legend_elements(), title="label")
ax.add_artist(legend1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```
